5. Generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap.

```python
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.rand(100),
    'B': np.random.rand(100),
    'C': np.random.rand(100)
})
data['D'] = 0.5 * data['A'] + 0.2 * data['B'] + np.random.normal(0, 0.02, 100)

corr = data.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix of Synthetic Data')

plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Data generation
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data for DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

# Update layout
fig.update_layout(
    title='3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh in Jupyter Notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', toolbar_location=None, tools='')

p.vbar(x=fruits, top=counts, width=0.4, color='navy', alpha=0.5)

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None

show(p)
```

```python


# Enable Bokeh in Jupyter Notebook
output_notebook()

# Generate histogram data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a histogram plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', tools='', toolbar_location=None)

# Add histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.7)

show(p)
```
